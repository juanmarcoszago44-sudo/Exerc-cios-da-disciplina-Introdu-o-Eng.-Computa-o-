```python
quantidade = int(input("Quantos números deseja digitar? "))

soma = 0

for i in range(quantidade):

    numero = int(input("Digite um número: "))

    if i == 0:
        maior = numero
        menor = numero

    if numero > maior:
        maior = numero

    if numero < menor:
        menor = numero

    soma = soma + numero

print("Maior:", maior)
print("Menor:", menor)
print("Soma:", soma)
```

```python
quantidade = int(input("Quantos números deseja digitar? "))

soma = 0

for i in range(quantidade):

    numero = int(input("Digite um número entre 0 e 1000: "))

    while numero < 0 or numero > 1000:
        print("Número inválido")
        numero = int(input("Digite novamente: "))

    if i == 0:
        maior = numero
        menor = numero

    if numero > maior:
        maior = numero

    if numero < menor:
        menor = numero

    soma = soma + numero

print("Maior:", maior)
print("Menor:", menor)
print("Soma:", soma)
```

```python
N = int(input("Digite um número: "))

divisoes = 0

for numero in range(1, N + 1):

    contador = 0

    for i in range(1, numero + 1):
        divisoes = divisoes + 1

        if numero % i == 0:
            contador = contador + 1

    if contador == 2:
        print(numero)

print("Quantidade de divisões:", divisoes)
```

```python
quantidade = int(input("Quantas pessoas? "))

soma = 0

for i in range(quantidade):
    idade = int(input("Digite a idade: "))
    soma = soma + idade

media = soma / quantidade

print("Média:", media)

if media <= 25:
    print("Turma jovem")

elif media <= 60:
    print("Turma adulta")

else:
    print("Turma idosa")
```

```python
eleitores = int(input("Quantidade de eleitores: "))

candidato1 = 0
candidato2 = 0
candidato3 = 0

for i in range(eleitores):

    voto = int(input("Vote no candidato 1, 2 ou 3: "))

    if voto == 1:
        candidato1 = candidato1 + 1

    elif voto == 2:
        candidato2 = candidato2 + 1

    elif voto == 3:
        candidato3 = candidato3 + 1

print("Votos candidato 1:", candidato1)
print("Votos candidato 2:", candidato2)
print("Votos candidato 3:", candidato3)
```

```python
turmas = int(input("Quantidade de turmas: "))

total_alunos = 0

for i in range(turmas):

    alunos = int(input("Quantidade de alunos da turma: "))

    while alunos > 40:
        print("A turma não pode ter mais de 40 alunos")
        alunos = int(input("Digite novamente: "))

    total_alunos = total_alunos + alunos

media = total_alunos / turmas

print("Média de alunos por turma:", media)
```
